```python
import pandas as pd
```

```python
#load the csv file of dataset into dataframe
f1=pd.read_csv("Data\PostComputationData.csv")
f1
```

Feature computation

```python
#creating the new attribute POM_Recharge
f1['POM_Recharge'] = (f1['PREMON']).astype(float) - (f1['MONSOON']).astype(float)

#print the updated dataframe
f1
```

```python
#creating the new attribute RB_Effect
f1['RB_Effect'] = (f1['POMKH']).astype(float) - (f1['POMRB']).astype(float)

#print the updated dataframe
f1
```

```python
#creating the new attribute KH_Effect
f1['KH_Effect'] = (f1['MONSOON']).astype(float) - (f1['POMKH']).astype(float)

#print the updated dataframe
f1
```

```python
f1.to_csv("Data\PostComputationData.csv",index=False,encoding='utf-8-sig')
```
