Q- What is the effect of change in number of wells in an area to its average grounwater level?

```python
import pandas as pd
import matplotlib.pyplot as plt
```

```python
df=pd.read_csv("Data/PostComputationData.csv")
```

```python
Years=[1996, 1997, 1998, 1999, 2000, 2001, 2002, 2003, 2004, 2005, 2006, 2007, 2008, 2009, 2010, 2011, 2012, 2013, 2014, 2015, 2016 ]
```

```python
no_of_wells=[]
for year in Years:
    subsetdf= df[df["YEAR_OBS"]==year]
    no_of_wells.append(len(subsetdf.index))
print(no_of_wells)
```

```python
GWL=[]
for year in Years:
    subsetdf= df[df["YEAR_OBS"]==year]
    GWL.append(subsetdf["MONSOON"].mean())
print(GWL)
```

```python
no_of_wells.so
plt.plot(no_of_wells, GWL, color='b')
plt.xlabel('Number of Wells')
plt.ylabel('Ground Water Level')
plt.title('Number of Wells to Ground Water Level')
plt.show()
```
